**Eigenvalues (explained_variance_)**

```python
model = PCA(n_components=6)
model.fit(df)

pd.DataFrame(model.explained_variance_ratio_, index=['y1', 'y2', 'y3', 'x1', 'x2', 'x3']).T
```

explained_variance_ratio - повертає матрицю, де елементи - відсоток дисперсії пояснений кожною компонентою.

Останні 3 компоненти найменше описують нашу вибірку, і це насправді було очікуванно, бо рівень глюкози в крові не стабілізується за одну годину, а під час посту рівень цукру стабільний і похибка вимірювання мінімальна.
